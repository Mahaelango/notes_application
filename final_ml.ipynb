```python
!pip -q install faiss-cpu
import pandas as pd
import numpy as np
import faiss
```

```python
data = pd.read_csv(r'/content/drive/MyDrive/amazon.csv')
data
```

```python
#taking only the necessary columns
data = data[['product_name', "category","rating","review_title"]].drop_duplicates().reset_index(drop=True)
data
```

```python
data.isnull().any()
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
data['all_details'] = data['product_name'] + " " + data['category'] +" " + data['rating'] +" " + data['review_title']
```

```python
data
```

```python
course_corpus = data['all_details']
```

```python
course_corpus
```

```python
  from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#vectorizer
vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(3,3), min_df=5)
X = vectorizer.fit_transform(course_corpus)
X = np.float32(X.toarray())
```

```python
X
```

```python
index = faiss.IndexFlatL2(X.shape[1])
index.add(X)
```

```python
ip=input("Enter the product name: ")
```

```python
ip
```

```python
search_text = [f"{ip}"]
search_text_vector = vectorizer.transform(search_text)
search_text_vector_array = np.float32(search_text_vector.toarray())

distances, indices = index.search(search_text_vector_array, 5)
```

```python
indices[0]
```

```python
#fetching names
product_names = [data['product_name'][indices[0][i]] for i in range(len(indices[0]))]
category_names = [data['category'][indices[0][i]] for i in range(len(indices[0]))]
rating_names = [data['rating'][indices[0][i]] for i in range(len(indices[0]))]
review_names = [data['review_title'][indices[0][i]] for i in range(len(indices[0]))]
```

```python
#converted into dataframes
product_names=pd.DataFrame(product_names)
category_names=pd.DataFrame(category_names)
rating_names=pd.DataFrame(rating_names)
review_names=pd.DataFrame(review_names)
```

```python
final = pd.concat([product_names, category_names, rating_names, review_names], axis=1)

# Rename the columns
final.columns = ['product_name', 'category', 'rating', 'review_title']
```

```python
final
```

```python
# Sort the DataFrame by 'rating' column in descending order
final = final.sort_values(by='rating', ascending=False)

```

```python
final
```

```python
from textblob import TextBlob

final['sentiment'] = final['review_title'].apply(lambda x: 'Positive review' if TextBlob(x).sentiment.polarity > 0 else 'Negative review')
```
